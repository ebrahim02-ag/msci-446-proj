## Writing Function to cross validate
ChatGPT was used in part to write this function, its general purpose is t

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import recall_score
from sklearn.metrics import make_scorer
def cross_validate_scoring(model,  X, y,model_name):
    """
    function to cross validate model over four different metrics
    model: Estimator Object
    X: features
    y: categorical
    model_name: string with the name of the model
    """
    scoring = ["accuracy","precision_macro","recall_macro","f1_macro"]
    scores = cross_validate(model, X, y, cv=10, scoring=scoring,return_estimator=True)
    
    print(f"{model_name}\nscores:{scores}")
    #most accurate
    best_scores = {
        "accuracy": round(scores["test_accuracy"].mean(),4),
        "precision_macro":round(scores["test_precision_macro"].mean(),4),
        "recall_macro":round(scores["test_recall_macro"].mean(),4),
        "f1_macro":round(scores["test_recall_macro"].mean(),4)
        

    }
    estimator = scores["estimator"]
    return best_scores, estimator

```

## Initialize models
Initialized 6 models 

Logistic Regression
Gaussian Naive Bayes
K Nearest Neighbors
XGBoost (gradient boosted classifier)
SVC: support vector classifier

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC

lr = LogisticRegression()
knc = KNeighborsClassifier()
gn = GaussianNB()
svc =  SVC()
rfc = RandomForestClassifier()
xgb = XGBClassifier()
```

```python
models = {
    "Logistic Regression"  : lr,
    "KNeighbors" : knc,
    "Gaussian Naive Bayes" : gn,
    "Random Forest" : rfc,
    "XGBoost": xgb,
    "SVC" : svc
    
}
```

## Generate metrics

```python
metrics = {}
trained_models = {}
for key, val in models.items():
    scores,model = cross_validate_scoring(model=val,X=X, y=y,model_name=key)
    metrics[key] = scores
    trained_models[key] = model
    
```
