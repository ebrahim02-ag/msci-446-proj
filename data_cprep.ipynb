```python
import csv
import pandas as pd
import numpy as np
```

```python
  # 'CC_action', 
  # 'CC_not_mentioned', 
  # 'CC_unkown',
  # 'CC_threat', 
  # 'CC_relative_impact', 

columns = [
  'rowID', 
  'speciesID',
  'gic_decision', 
  'Critical_habitat',
  'sara_status', 
  '_threat_identified', 
  'iucn_impact', 
  'iucn_scope', 
  'iucn_severity',
  'iucn_timing',  
]
```

```python
df = pd.read_csv('./CAN-SAR_database.csv',  encoding='ISO-8859-1', index_col=0)
```

```python
filtered_df = df[[col for col in df.columns if any(item in col for item in columns)]].copy()
```

```python
filtered_df
```

```python
Status = {
  'Endangered': 0,
  'Threatened': 1,
  'Special Concern': 2
}

filtered_df['sara_status'] = filtered_df['sara_status'].map(Status)
filtered_df.dropna(subset=['sara_status'], inplace=True)
filtered_df.value_counts('sara_status')
```

```python
Crtical_habitat = {
  'N': 0,
  'Y': 1,
  'P': 2, 
  np.nan: -1,
}

filtered_df['Critical_habitat'] = filtered_df['Critical_habitat'].map(Crtical_habitat)
filtered_df.value_counts('Critical_habitat')
```
